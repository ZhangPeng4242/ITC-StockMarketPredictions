```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import yfinance as yf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df = yf.download('DIDI', interval='1d')["Adj Close"]
```

```python
df.plot()
```

```python
df.diff().plot()
```

```python
from statsmodels.tsa.arima.model import ARIMA, ARMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(df.diff(1).dropna());
```

```python
plot_pacf(df.diff(1).dropna());
```

```python
iloc  = int(np.floor(len(df.index)*0.8))
iloc
```

```python
X_train = df.iloc[:iloc]
X_test = df.iloc[iloc:]
X_train, X_test
```

```python
arima = ARIMA(X_train,order=(0,1,0))
arima_fit = arima.fit()
pred_arima = arima_fit.forecast()
```

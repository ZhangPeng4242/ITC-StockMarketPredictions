```python
import numpy as np
import pandas as pd
df=pd.read_csv('transactions_bestsect.csv')
```

```python
df.head(50)
```

```python
df['Positive_return']=np.select([df['profit']>0 ,(df['profit']==0) & (df.type=='buy'),
                                 (df['profit']==0) & (df.type=='sell'),df['profit']<0],[1,None,-1,-1])
```

```python
df.Positive_return.value_counts(dropna=False)
```

```python
df.loc[ (df.Positive_return!=1)&(df.Positive_return!=-1) ].type.value_counts()
```

```python
tickers=df.ticker.unique().tolist()
```

```python
tickers
```

```python
occurences=[]
tick_prof=[]
avg_tick_profit=[]
profitable=[]
for ticker in tickers:
    occurence=len(df.loc[df.ticker==ticker])/2
    occurences.append(occurence)
    tot_profit=round(df.loc[df.ticker==ticker].profit.sum(),1)
    tick_prof.append(tot_profit)
    avg_tick_profit.append(round(tot_profit/occurence,1))
    profitable.append(round(len(df.loc[(df.ticker==ticker)&(df.Positive_return>0)])/occurence,5)*100)
stats=pd.DataFrame()
stats['ticker']=tickers
stats['occurences']=occurences
stats['profit_made_from_ticker']=tick_prof
stats['average_profit made/investment']=avg_tick_profit
stats['Positive_returns/Nb_of_investments']=profitable
```
